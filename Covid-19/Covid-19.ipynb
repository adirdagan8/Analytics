```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotly.subplots as make_subplots
import plotly.express as px
import datetime as datetime
```

```python
covid_df = pd.read_csv('covid_19_india.csv')
covid_df.head(10)
```

```python
vaccine_df = pd.read_csv('covid_vaccine_statewise.csv')
vaccine_df.head(7)
```

```python
covid_df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],inplace=True,axis=1)
```

```python
covid_df.head()
```

```python
covid_df.Date = pd.to_datetime(covid_df.Date,format='%Y-%m-%d')
```

```python
covid_df
```

```python
covid_df['active_Cases'] = covid_df.Confirmed - (covid_df.Cured + covid_df.Deaths)
covid_df.head()
```

```python
statewise = pd.pivot_table(covid_df,values=['Confirmed','Deaths','Cured']
                           ,index='State/UnionTerritory',aggfunc=max)
```

```python
statewise['Recovery Rate'] = 100*statewise.Cured/statewise.Confirmed
```

```python
statewise['Mortality Rate'] = 100*statewise.Deaths/statewise.Confirmed
```

```python
statewise = statewise.sort_values(by= 'Confirmed',ascending=False)
```

```python
statewise.style.background_gradient(cmap= 'cubehelix')
```

```python
# Top 10 active cases states

Top_10_active_cases = covid_df.groupby(by = 'State/UnionTerritory').max()[['active_Cases','Date']].sort_values(by = ['active_Cases'],ascending = False).reset_index()
```

```python
fig= plt.figure(figsize=(16,9))
plt.title('Top 10 states with most active cases in india',size=25)
ax = sns.barplot(data=Top_10_active_cases.iloc[:10],x='State/UnionTerritory',y='active_Cases',linewidth = 2,edgecolor='red')
plt.xlabel('States',size=20)
plt.ylabel('Total active case',size=20)
plt.show()
```

```python
# Top states with highest deaths

Top_10_active_cases = covid_df.groupby(by = 'State/UnionTerritory').max()[['Deaths','Date']].sort_values(by = ['Deaths'],ascending = False).reset_index()

fig= plt.figure(figsize=(18,5))
plt.title('Top 10 states with most deaths',size=25)
ax = sns.barplot(data=Top_10_active_cases.iloc[:12],x='State/UnionTerritory',y='Deaths',linewidth = 2,edgecolor='black')
plt.xlabel('States',size=20)
plt.ylabel('Total Deaths cases',size=20)
plt.show()
```

```python
# Growth trend

fig= plt.figure(figsize=(12,6))
ax = sns.lineplot(data=covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Kerala','Tamil Nadu','Uttar Pradesh'])],x='Date',y='active_Cases',hue='State/UnionTerritory')
ax.set_title('Top 5 Affected States in India',size=25)
plt.xlabel('States',size=20)
plt.ylabel('Total Deaths cases',size=20)
plt.show()
```

```python
#Remove rows where state = India

vaccine = vaccine_df[vaccine_df.State!='India']
vaccine
```

```python
vaccine.rename(columns={'Total Individuals Vaccinated':'Total'},inplace=True)
vaccine.head()
```

```python
#Most vaccinated State

max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total', ascending=False)[:5]
max_vac
```

```python
fig= plt.figure(figsize=(10,5))
plt.title('Top 5 Vaccinated States in india',size=20)
ax = sns.barplot(data=max_vac,x=max_vac.index,y='Total',linewidth = 2,edgecolor='black')
plt.xlabel('States',size=20)
plt.ylabel('Vaccination',size=20)
plt.show()
```

```python
#Most vaccinated State

min_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
min_vac = min_vac.sort_values('Total', ascending=True)[:5]
min_vac
```

```python
fig= plt.figure(figsize=(16,9))
plt.title('Top 5 Vaccinated States in india',size=20)
ax = sns.barplot(data=min_vac,x=min_vac.index,y='Total',linewidth = 2,edgecolor='black')
plt.xlabel('States',size=20)
plt.ylabel('Vaccination',size=20)
plt.xticks(rotation=30,size=15)
plt.yticks(size=15)
plt.show()
```
