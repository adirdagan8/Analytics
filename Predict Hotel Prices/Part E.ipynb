```python
# Define the merge sort algorithm for sorting rows of the DataFrame
def merge_sort_df(df, sort_columns):
    if len(df) <= 1:
        return df
    
    # Divide the DataFrame in half
    mid = len(df) // 2
    left_half = df[:mid]
    right_half = df[mid:]
    
    # Recursively sort the left and right halves
    left_half = merge_sort_df(left_half, sort_columns)
    right_half = merge_sort_df(right_half, sort_columns)
    
    # Merge the two sorted halves
    merged = pd.concat([left_half, right_half])
    merged = merged.sort_values(by=sort_columns)
    
    return merged

# Create a sample DataFrame with multiple columns
df = pd.DataFrame({
    'Column1': [2, 1, 3],
    'Column2': [4, 5, 6],
    'Column3': ['A', 'B', 'C'],
    'Column4': ['X', 'Y', 'Z']
})

# Sort the rows of the DataFrame using the merge sort algorithm based on multiple columns
sort_columns = ['Column1', 'Column2', 'Column3', 'Column4']
sorted_df = merge_sort_df(df, sort_columns)

# Print the resulting DataFrame
print(sorted_df)
```

```python
#LinearRegression

#Function to find RMSE MSE MAE R2

def reg_metrics(y_test, y_pred, x_train):
    
    #imports
    from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error


    #calculate
    RMSE = np.sqrt(mean_squared_error(y_test,y_pred))
    MSE = mean_squared_error(y_test,y_pred)
    MAE = mean_absolute_error(y_test,y_pred)
    R2 = r2_score(y_test,y_pred)
    
    #print
    print("RMSE: " + str(round(RMSE,3)))
    print("MSE:  " + str(round(MSE,3)))
    print("MAE:  " +  str(round(MAE,3)))
    print("R2:   " + str(round(R2,3)))



#test and train
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)

# Create model
from sklearn import linear_model
reg = linear_model.LinearRegression()
from matplotlib import pyplot


# Train the model
reg.fit(x_train, y_train)

# Make predictions
y_pred = reg.predict(x_test)


#Calculate
reg_metrics(y_test, y_pred, x_train)


#Actual vs Predict
plt.title('Predict vs Actual')
plt.xlabel("Actual Price")
plt.ylabel("Predict Price")
X = np.linspace(0,8000,100)
Y = 1*X+0
plt.plot(X, Y, '-r')
plt.scatter(y_pred,y_test,alpha=0.25)
plt.xticks(np.arange(0,10000,step=1000))
plt.yticks(np.arange(0,10000,step=1000))
plt.show()




# get importance
plt.title('importance')
importance = reg.coef_
# summarize feature importance
pyplot.bar([x.columns[j] for j in range(len(importance))], importance)
pyplot.xticks(fontsize = 9,rotation=45)
pyplot.show()





import statsmodels.api as sm

#add constant to predictor variables
x_r = sm.add_constant(x)

#fit linear regression model
model = sm.OLS(y, x_r).fit() 


#create instance of influence
influence = model.get_influence()

#obtain standardized residuals
standardized_residuals = influence.resid_studentized_internal



plt.scatter(bookingHotels['Price'], standardized_residuals,alpha=0.35)
plt.xlabel('x')
plt.ylabel('Standardized Residuals')
plt.axhline(y=0, color='black', linestyle='--', linewidth=1)
plt.show()
```
